```python
from deeppavlov.core.commands.utils import parse_config

model_config = parse_config('ner_collection3_bert')

# dataset that the model was trained on
print(model_config['dataset_reader']['data_path'])
```

```python
new_tags = [
    "O",
    'B-Дата',
    'B-бренд',
    'B-вид спорта',
    'B-видеоигра',
    'B-команда',
    'B-лига',
    'B-локация',
    'B-модель',
    'B-название проекта',
    'B-организация',
    'B-персона',
    'B-сезон',
    'B-серия',
    'I-Дата',
    'I-бренд',
    'I-вид спорта',
    'I-видеоигра',
    'I-команда',
    'I-лига',
    'I-локация',
    'I-модель',
    'I-название проекта',
    'I-организация',
    'I-персона',
    'I-сезон',
    'I-серия'
]
```

```python
new_tags_txt = "\t0\n".join(new_tags) + "\t0"
# new_tags_txt = new_tags_txt[:-1]
```

```python
print(new_tags_txt)
```

```python
with open("our_tags.dict", 'w') as f:
    f.write(new_tags_txt)
```

```python
model_config["chainer"]["pipe"][1]["freq_drop_load"] = True
```

```python
model_config["chainer"]["pipe"][1]["save_path"] = "/content/our_tags_saved.dict"
model_config["chainer"]["pipe"][1]["load_path"] = "/content/our_tags.dict"

# to be compatible with original model
model_config["chainer"]["pipe"][2]["n_tags"] = 13
```

```python
from deeppavlov import build_model

ner_model = build_model(model_config, download=True, install=False)
```

```python
from torch import nn
```

```python
bert = ner_model[2].model
```

```python
# replace last layer to work with our tags
bert.classifier = nn.Linear(768, len(new_tags), bias=True) #.cuda()
bert.num_labels = len(new_tags)
model_config["chainer"]["pipe"][2]["n_tags"] = len(new_tags)
```
